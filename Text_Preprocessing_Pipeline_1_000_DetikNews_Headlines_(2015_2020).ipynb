```python
df = pd.read_csv("scrapedDetikNEWS2015-2020.csv")
df.head()
```

```python
df_clean = df.iloc[3000:4000].copy()
df_clean.reset_index(drop=True, inplace=True)
```

```python
df_clean.dropna(inplace=True)
df_clean.drop_duplicates(inplace=True)
```

```python
df_clean.shape
```

```python
df_clean.head()
```

# **Preprocessing**

# Lowercase

```python
df_clean["lowercase"] = df_clean["name"].str.lower()
df_clean.head()
```

# Cleaning

```python
def clean_text(text):
  # Hapus semua karakter kecuali huruf alfabet dan spasi
  text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)

  # Hapus spasi berlebih dengan satu spasi
  text = re.sub(r'\s+', ' ', text)

  # Hilangkan spasi di awal dan akhir
  text = text.strip()

  return text

df_clean["cleaned"] = df_clean["lowercase"].apply(clean_text)
df_clean[["name", "lowercase", "cleaned"]].head()
```

# Stopwords Removal

```python
factory = StopWordRemoverFactory()
stopwords_remover = factory.create_stop_word_remover()

def remove_stopwords(text):
    return stopwords_remover.remove(text)

df_clean["stopwords"] = df_clean["lowercase"].apply(remove_stopwords)

df_clean.head()
```

# Tokenized

```python
df_clean["tokenized"] = df_clean["stopwords"].apply(lambda x: x.split())
df_clean.head()
```

# Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tokens):
    return " ".join([stemmer.stem(w) for w in tokens])

df_clean["stemmed"] = df_clean["tokenized"].apply(stemming)
df_clean.head()
```
